```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#reading the dataset
ap_data = pd.read_csv('https://raw.githubusercontent.com/analyticsindiamagazine/MocksDatasets/main/AirPassengers.csv',index_col = "Month")
```

```python
ap_data.head()
```

```python
ap_data.shape
```

```python
ap_data.isnull().sum()
```

```python
ap_data.columns = ['No. of Passengers']
```

```python
plt.rcParams["figure.figsize"]=(15,10)
ap_data.plot().set_title("Yearly Number of Air Passengers")
plt.show()
```

```python
decompose_data = seasonal_decompose(ap_data, model="additive", period=30)
```

```python
plt.figure(figsize=(15,7))
level = decompose_data.observed
level.plot().set_title("Level")
```

```python
plt.figure(figsize=(15,7))
seasonality = decompose_data.seasonal
seasonality.plot().set_title("Seasonality")
```

```python
plt.figure(figsize = (15,7))
decompose_data.trend.plot().set_title("Trend")
```

```python
plt.figure(figsize = (15,7))
residual = decompose_data.resid
residual.plot().set_title("Residual")
```
